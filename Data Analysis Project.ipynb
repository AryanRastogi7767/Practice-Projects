```python
data['Subtitle'].isnull().value_counts()
```


**As "Subtitle" column has 11746 null values out of 17007, it will not be of much use in analysis and can be dropped.**


```python
data.drop(['Subtitle'],axis = 1,inplace = True)
```

```python
data.head()
```

**The "Languages" Column has few null values and therefore can be filled with 'EN' as it is one of the most popular language in the world.**

```python
data['Languages'].fillna(value = 'EN',inplace = True)
```

```python
sns.heatmap(data.isnull())
```

**Now removing null values of the price column by replacing them with mean of the values in the column.**

```python
mean_price=np.mean(data['Price'])
```

```python
data['Price'].fillna(value= mean_price,inplace = True)
```

**Now dealing with the null value in 'In-app Purchases' column.**

```python
data['In-app Purchases'].fillna(value='Unknown',inplace = True)
```

```python
sns.heatmap(data.isnull())
```

```python
d = data[data['Average User Rating'].isnull()==False]
```

```python
d.reset_index(inplace=True)
d.head()
```

```python
sns.heatmap(d.isnull())
```

**Now we can see that our data does not have much null values left. So we can use this cleaned data for furthur analysis.**

```python
lang_list=[]
for i in d['Languages']:
    l=i.split(',')
    for j in range(len(l)):
        lang_list.append(l[j])
```

```python
langs = set(lang_list)
re = set(['AF','AR','BE','BG','CA','CS','DA','EN','ES','FR','ID','IT','JA','KO','NL','PT','RU','SQ','TR','ZH'])
nlangs = langs - re
```

```python
c = {}
for i in langs:
    count = 0
    for j in lang_list:
        if i==j:
            count+=1
    c.update({i:count})        
c
```

**From this dictionary we can see that "FR, ES, RU, EN, IT, DE, JA, ZH" are some of the most popular lanuages in the data.**

```python
month =[]
for i  in range(len(d['Original Release Date'])):
    a = int(d['Original Release Date'][i].split('/')[1])
    month.append(a)
```

```python
month = pd.DataFrame(month)
d = pd.concat([d,month],axis=1)
```

```python
sns.countplot(d[0])
plt.tight_layout()
```

**From this we can see that games released in months like March, April, May and December have comparatively higher user rating.**

**In order to get a good user rating from a strategy game the best combination would be:**

1. The size of the game close to  162894622.200 would be preferred.
